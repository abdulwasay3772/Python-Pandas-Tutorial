```python
## how to parse date
## convert it to proper datetime format
## year,month,day, space, (I for 12 Hour clock), (p for local AM or PM)
df['Date']=pd.to_datetime(df['Date'], format='%Y-%m-%d %I-%p')
```

```python
## now after parsing 
## day_name on that date
df.loc[0,'Date'].day_name()
```

```python
## second method to convert to proper foramt
## lambda is the anonymous function
## strptime (convert string to datetime objects)
d_parser= lambda x: pd.datetime.strptime(x,'%Y-%m-%d %I-%p')
```

```python
## To read the date column correctly, we can use the argument parse_dates to specify a list of date columns
df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df.head()
```

```python
## again for day_name
df.loc[0, 'Date'].day_name()
```

```python
df['Date']
```

```python
## day_name of entire date column
df['Date'].dt.day_name()
```

```python
## create column ---> display the day of week 
df['Day_Of_Week'] = df['Date'].dt.day_name()
```

```python
df
```

```python
## earliest date
df['Date'].min()
```

```python
## most recent date
df['Date'].max()
```

```python
## time(Dalta)=subtract these two dates
## tells about the data period(data of how many days)
df['Date'].max() - df['Date'].min()
```

```python
## data of the year 2020 and onwards
filt = (df['Date'] >='2020')
df.loc[filt]
```

```python
## data between 2019-01-01 and 2020-01-01
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt]
```

```python
## set index to date
## inplace --> to change permanently
df.set_index('Date', inplace=True)
```

```python
## high for that day
## from resample we get high value of each day of data
## D for day
## W for week etc
highs = df['High'].resample('D').max()
highs['2020-01-01']
```

```python
## to plot the data
## inline use to display the plot with in the browser
%matplotlib inline
```

```python
## high value plot
highs.plot()
```

```python
## resample entire data to week 
## display the weekly data
a=df.resample('W').mean()
```

```python
a
```

```python
a.plot()
```

```python
## weekly aggregated data
b=df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'})
```
