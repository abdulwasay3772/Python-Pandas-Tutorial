Read and write data to different sources

```python
import pandas as pd
```

```python
## read the data
df=pd.read_csv ('survey_results_public.csv')
## to read the seted file
schema_df= pd.read_csv('survey_results_schema.csv')
```

```python
## set option use for to display columns which we wanna see
pd.set_option('display.max_columns',85)
## set option use for to display rows which we wanna see
pd.set_option('display.max_rows',85)
```

```python
df.head(5)
```

```python
## filter the pakistan in country column
filt = (df['Country'] == 'Pakistan')
```

```python
## create new dataframe and pass into filter
pakistan_df = df.loc[filt]
pakistan_df.head()
```

```python
## export the filtered data to csv file
pakistan_df.to_csv('modified.csv')
```

```python
## export the file in another format
pakistan_df.to_csv('modified.tsv', sep='\t')
```

```python
## export to excel xlsx file
pakistan_df.to_excel('modified.xlsx')
```

```python
## read created excel file
## set index to respondent column
test = pd.read_excel('modified.xlsx', index_col='Respondent')
```
