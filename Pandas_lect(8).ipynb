```python
## how many people respondent on survey from each coutry
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
## from each 'country' how many people know 'Python; 
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
## 'concat' the 2 series 
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
## rename the column
python_df.rename(columns={'Country': 'Num_Respondents', 'LanguageWorkedWith': 'Num_KnowsPython'}, inplace=True)
python_df
```

```python
## % knows python
python_df['Pct(%)_KnowsPython'] = (python_df['Num_KnowsPython']/python_df['Num_Respondents']) * 100
python_df
```

```python
## % by sorting
python_df.sort_values(by='Pct(%)_KnowsPython', ascending=False, inplace=True)
```
