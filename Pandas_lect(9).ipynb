How to Deal with missing data

```python
import pandas as pd
import numpy as np
```

```python
## Create a Dictionary
people = {
    "first": ['Abdul', 'Shah','Adnan','Basharat',np.nan,None,'NA'], 
    "last": ['Wasay', 'Khan', 'Khan','Ali',np.nan, np.nan, 'Missing'], 
    "email": ['abdulwasay@gmail.com', 'shahkhan@gmail.com', 'adnankhan@gmail.com', None, np.nan, 'Anonymous@gmail.com', 'NA'],
    "age": ['21', '23', '23', '22', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
## drop missing values 
df.dropna()
```

```python
## drop na values by index(Rows) 
## how=all if there is all values in row is missing 
df.dropna(axis='index', how='all')
```

```python
## drop na values by index(Rows) 
## how=any if there is any values in row is missing 
df.dropna(axis='index', how='any')
```

```python
## drop na values by columns
## how=all if there is all values in columns is missing 
df.dropna(axis='columns', how='all')
```

```python
## drop na values by columns
## how=any if there is any value in column is missing 
## b/c in all columns there is some missing value
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='index', how='all', subset=['email'])
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
## drop the row contains missing vale in last name column or email column
df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
## mask the vlaues(na or not)
df.isna()
```

```python
## fill na value=0
df.fillna(0)
```

```python
## dataframe datatype
## all of the columns are object
df.dtypes
```

```python
## error due to 
## column is string not integer datatype
df['age'].mean()
```

```python
## np.nan has float datatype
type(np.nan)
```

```python
## if column contain missing values then pass float datatype
df['age'] = df['age'].astype(float)
```

```python
df.dtypes
```

```python
## now mean of age
df['age'].mean()
```
